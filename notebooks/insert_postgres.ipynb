```python
cur = conn.cursor()


def get_meal_id(meal_name):
    cur.execute(f"SELECT id FROM meal WHERE name like '{meal_name}'")
    output = cur.fetchone()
    if output is not None:
        return output[0]
    else:
        return None


meal_id_1 = get_meal_id("Default Meal 1")
meal_id_2 = get_meal_id("Default Meal 2")
meal_id_3 = get_meal_id("Default Meal 3")
meal_id_4 = get_meal_id("Default Meal 4")
meal_id_5 = get_meal_id("Default Meal 5")

cur.close()
```
